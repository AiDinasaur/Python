```python
a = 0

while a<5: # infite for loop , because the a value is always less than 5
    print("Python")
```

```python
a = 0

while a<5:
    a = a+1
    print(a)
```

```python
a = 0

while a<5:
    a = a+1
    print(a)
else:
    print("Loop is Over")
```

```python
a = 0

while a<5:
    a = a+1
    if a==3:
        break
    print(a)
else:                     #if while loop ends prematurely the loop ends it won't execute the else block 
    print("Loop is Over")
```

```python
a = 0

while a<5:
    a = a+1
    if a==3:
        continue # skips at 3
    print(a)
else:                     #if while loop ends prematurely the loop ends it won't execute the else block 
    print("Loop is Over")
```

# Programs Using While Loop

```python
# Finding the number of digits in the number
```

```python
n = 153 
count = 0

while n>0:
    count = count +1
    n = n//10
print(count)
```

```python
# Finding the sum of digits in the number
```

```python
n = 153 
total = 0

while n>0:
    remainder = n%10
    total = total+remainder
    n = n//10

print(total)
```

```python
# Finding the reverse of digits in the number
```

```python
n = 153; 
rev = 0
  
while n>0: 
    a = n % 10
    rev = rev * 10 + a 
    n = n // 10
      
print(rev) 
```
