```python
#this is a program to display multiplication table
#example for nested-while loop in Python
i=1
while i<=10:
  j=1
  while j<=12:
        print(i*j)
        j+=1
  i+=1
  print("\n")
```

```python
#this is a program to display multiplication table
#example for nested while loop in Python
i=1
while i<=5:
  j=1
  while j<=5:
        print(i,"*",j,'=',i*j)
        j+=1
  i+=1
  print("\n")
```
